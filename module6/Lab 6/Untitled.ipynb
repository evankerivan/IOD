```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt

# Generate a simple dataset
np.random.seed(42)
X = np.random.randn(100, 2)
# Introduce some outliers
outliers = np.array([[8, 8], [9, 9], [10, 10]])
X_with_outliers = np.vstack([X, outliers])
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42)
```

```python
kmeans.fit(X_with_outliers)

labels = kmeans.labels_
```

```python
plt.scatter(X_with_outliers[:, 0], X_with_outliers[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='x')
plt.title("K-Means Clustering with Outliers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Apply K-Means without scaling & Plot without scaling


```python
kmeans.fit(X)

labels = kmeans.labels_
```

```python
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='x')
plt.title("K-Means Clustering with Outliers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Apply K-Means with scaling & Plot scaling


```python
>>> from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()  # Instantiate the scaler
scaled = scaler.fit_transform(X)  # Scale the data
print(scaled)
# Apply K-means clustering
kmeans = KMeans(n_clusters=2, random_state=42)  # Adjust the number of clusters as needed
kmeans.fit(scaled)

# Get the cluster labels
labels = kmeans.labels_

# Plot the results
plt.scatter(scaled[:, 0], scaled[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='x')
plt.title("K-Means Clustering with Outliers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
