```python
from PyPDF2 import PdfFileReader, PdfFileWriter

# Read the PDF
input_pdf_path = 'Invoice INV-0064 (1).pdf'
output_pdf_path = 'Invoice INV-0064.pdf'

reader = PdfFileReader(input_pdf_path)
writer = PdfFileWriter()

# Copy pages
for page_num in range(reader.numPages):
    writer.addPage(reader.getPage(page_num))

# Update metadata
metadata = reader.getDocumentInfo()
new_metadata = {
    '/Title': metadata.title if metadata.title else 'Untitled',
    '/Author': 'New Author',
    '/Subject': metadata.subject if metadata.subject else 'None',
    '/Producer': 'New Producer Software',
    '/Creator': 'New Content Creator Software',
    '/CreationDate': metadata['/CreationDate'] if '/CreationDate' in metadata else '',
    '/ModDate': metadata['/ModDate'] if '/ModDate' in metadata else ''
}

writer.addMetadata(new_metadata)

# Write to a new PDF
with open(output_pdf_path, 'wb') as out_file:
    writer.write(out_file)
```

```python
from PyPDF2 import PdfReader, PdfWriter

# Read the PDF
input_pdf_path = 'Invoice INV-0064 (1).pdf'
output_pdf_path = 'Invoice_INV-0064_updated.pdf'

reader = PdfReader(input_pdf_path)
writer = PdfWriter()

# Copy pages
for page_num in range(len(reader.pages)):
    writer.add_page(reader.pages[page_num])

# Update metadata
metadata = reader.metadata
new_metadata = {
    '/Title': metadata.get('/Title', 'Untitled'),
    '/Author': 'New Author',
    '/Subject': metadata.get('/Subject', 'None'),
    '/Producer': 'Aspose.Pdf for .NET 6.6',
    '/Creator': 'Aspose Ltd.',
    '/CreationDate': metadata.get('/CreationDate', ''),
    '/ModDate': metadata.get('/ModDate', '')
}

writer.add_metadata(new_metadata)

# Write to a new PDF
with open(output_pdf_path, 'wb') as out_file:
    writer.write(out_file)
```
