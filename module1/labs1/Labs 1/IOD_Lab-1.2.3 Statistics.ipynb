# 1. Plot a pie chart of the frequency of some observations

Modify the `explode=` line to "explode" only the third slice (i.e. 'Data Scientists')

````python
# frequency of some observations using pie chart
labels = 'Developers', 'Data Engineers', 'Data Scientists', 'Business Analysts'
frequency = [15, 32, 12, 7]
fig1, ax1 = plt.subplots()
explode = (0, 0, .25, 0)  # MODIFY THIS

ax1.pie(frequency, explode=explode, labels=labels, autopct='%.2f',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
````

# 2. Plot the frequency of some observations using a bar chart
